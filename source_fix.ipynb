```python
import os
import pandas as pd
import numpy as np
from datetime import date
from datetime import time
from datetime import datetime
```

```python

# Remove output files
output_directory = "C:/Users/rhett/NewsListsOutput"
output_files = os.listdir(output_directory)

for f in output_files:
    os.remove(os.path.join(output_directory, f))

# Get list of files in input directory
input_directory = "C:/Users/rhett/NewsLists/ChangeEventSource/"
input_files = os.listdir(input_directory)

# Sort the list of files 
sorted_files = sorted(input_files)

# Sort by est time
for f in sorted_files:
    file_path = os.path.join(input_directory, f)
    df = pd.read_csv(file_path)
    # Add col for eastern date and conversion
    df['timestamp2'] = df['timestamp']
    df['timestamp2'] = [datetime.utcfromtimestamp(x/(1000000)) for x in df['timestamp2']]
    df['timestamp2'] = pd.to_datetime(df['timestamp2'], unit='ms')
    df['timestamp2'] = pd.to_datetime(df['timestamp2']).dt.tz_localize('UTC').dt.tz_convert('US/Eastern')
    #df['date_et'] = df['timestamp'].dt.date
    df['date_et'] = df['timestamp2'].dt.date
    # Get distinct values from date_et col
    days = df['date_et'].unique()
    # Select rows where date_et = d
    for d in days:
        df2 = df.loc[df['date_et'] == d]
        df2 = df2.drop(['timestamp2','date_et'], axis=1)
        # Append df2 to output_file
        fname = d.strftime("%Y-%m-%d")
        #if file already exists dont include header
        path = "C:/Users/rhett/NewsLists/NewsListsOutput/%s.csv"%(fname)
        show_header = not os.path.isfile(path)
        df2.to_csv(path, mode='a', index=False, header=show_header)
df2
```
